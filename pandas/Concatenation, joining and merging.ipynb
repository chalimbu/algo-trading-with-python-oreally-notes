```python
import numpy as np
import pandas as pd
df1= pd.DataFrame(['100','200','300','400'],
                  index=['a','b','c','d'],
                  columns=['A',])
df1
```

```python
df2 = pd.DataFrame(['200','150','50'],
                  index=['f','b','d'],
                  columns = ['B',])
df2
```

```python
#do the same
df1.append(df2,sort=False)
pd.concat((df1,df2),sort=False)
```

```python
#do the same
df1.append(df2,ignore_index=True,sort=False)
pd.concat((df1,df2),ignore_index=True,sort=False)
```

```python
#the joins worth with the index
df1.join(df2)
```

```python
df2.join(df1)
```

```python
#default
df1.join(df2,how='left')
```

```python
#the same as df2.join(df1)
df1.join(df2,how='right')
```

```python
#in both
df1.join(df2,how='inner')
```

```python
#preserves all indixes
df1.join(df2,how='outer')
```

```python
df = pd.DataFrame()
df['A']=df1['A']
df
```

```python
df['B']=df2
df
```

```python
df = pd.DataFrame({'A': df1['A'], 'B': df2['B']})
df
```

```python
#mergin
c = pd.Series([250,150,50],index=['b','d','c'])
df1['C']=c
df2['C']=c
df1
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,left_on='A',right_on='B')
```

```python
pd.merge(df1,df2,left_on='A',right_on='B',how='outer')
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,on='C',left_index=True)
```

```python
pd.merge(df1,df2,on='C',right_index=True)
```
